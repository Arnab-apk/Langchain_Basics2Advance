### First Generative AI Application

```python
from openai import base_url
from langchain.chat_models import init_chat_model
import os

model = init_chat_model(
    model="gpt-5",
    base_url="https://openrouter.ai/api/v1",
    api_key=os.environ.get("OPENAI_API_KEY")
    
)
response=model.invoke("write me a simple python code to print hello world")
print(response.content)
```

```python
from langchain_openai import ChatOpenAI
import os
# Initialize the model pointing to OpenRouter
model = ChatOpenAI(
    model="openai/gpt-5", # Or any model supported by OpenRouter
    api_key=os.environ.get("OPENAI_API_KEY"),
    base_url="https://openrouter.ai/api/v1"
)
# Now you can use the model
response = model.invoke("Write me an essay on AI")
print(response.content)
```

### LLM Model Integration With Google Gemini

```python
os.environ["GEMINI_API_KEY"] = os.getenv("GEMINI_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model=init_chat_model("google_genai:gemini-2.5-flash-lite")
response=model.invoke("Hello, how are you?")
print(response.content)
```

### Grok Langchain Integration

```python
from langchain.chat_models import init_chat_model 
import os  
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
    
```

```python
model = init_chat_model("llama-3.1-8b-instant", model_provider="groq")
response = model.invoke("Write me a joke on ML algorithms")
print(response.content)
```

```python
import os
from langchain_groq import ChatGroq

# Make sure GROQ_API_KEY is in your environment or .env
model = ChatGroq(
    model="qwen/qwen-3-32b",
    api_key=os.environ.get("GROQ_API_KEY")
)

response = model.invoke("Why do parrots speak?")
print(response.content)
```
