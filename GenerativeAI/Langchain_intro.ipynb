### Streaming And Batch

Streaming, most model can stream their output content while its being generated. By displaying their output progresseively, streaming significantly improves teh use experience, pirticularly for longer responses. Calling a stream() returns an iterator that yeids output as chunkks as teh are being generated. You can also use a loop to process each chunk in real time

```python
for chunk in model.stream("Write me a 1000 word essay on AI"):
    print(chunk.text,end="",flush=True)
```

```python
for chunk in model.stream("Why does parrots have colourful featheres?"):
    print(chunk.text,end="",flush=True)
```

### Batch 
Batching is a collection of independent requests to a model can significantly improve performance and reduce costs as thr processing can be done in parallel.

```python
responses=model.batch(
    ["why do parrots have coloourful feathers?",
    "How do aeroplanes fly?",
    "What is quantum computing?"]
)

for response in responses:
    print(response.content)
```

```python
responses=model.batch(
    ["why do parrots have coloourful feathers?",
    "How do aeroplanes fly?",
    "What is quantum computing?"],
    config={
        'max_concurrency':4,#Limit to 4 parallel calls
    }
)

for response in responses:
    print(response)
```
